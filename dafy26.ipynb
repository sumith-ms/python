```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import numpy as np
```

```python
cars = pd.read_csv(r"/content/Cars.csv")
cars.head()
```

```python
cars.info()
```

```python
cars.isna().sum()
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(cars['HP'])
plt.title('HP Distribution Plot')
```

```python
# prompt: boxplot in  python

import matplotlib.pyplot as plt
plt.subplot(1,2,2)
sns.boxplot(cars['SP'])
plt.title('SP Box Plot')
plt.show()
```

```python
# prompt: defirnd sactterplot

import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.scatterplot(x='HP',y='SP',data=cars)
plt.title('Scatter Plot of HP vs SP')
plt.show()
```

```python
cars.corr()
```

```python
import matplotlib.pyplot as plt
for col in cars.columns:
  if col != 'HP':
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='HP', y=col, data=cars)
    plt.title(f'Scatter Plot of HP vs {col}')
    plt.xlabel('HP')
    plt.ylabel(col)
    plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.scatterplot(x='VOL', y='SP', data=cars)
plt.title('Scatter Plot of VOL vs SP')
plt.xlabel('VOL')
plt.ylabel('SP')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.scatterplot(x='VOL', y='WT', data=cars)
plt.title('Scatter Plot of VOL vs WT')
plt.xlabel('VOL')
plt.ylabel('WT')
plt.show()
```
